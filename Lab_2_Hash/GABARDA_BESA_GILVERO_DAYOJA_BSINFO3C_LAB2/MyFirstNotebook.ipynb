```python
# Cell 1: Import required libraries
import hashlib
import pandas as pd
import numpy as np
from datetime import datetime
import json
import os
import warnings
warnings.filterwarnings('ignore')

# For better display in Jupyter
from IPython.display import display, HTML, Markdown

print("✅ Libraries imported successfully!")
```

# 2nd

```python
# Cell 2: Integrity Guardian Class Definition
class IntegrityGuardian:
    """
    A tool to monitor file integrity using cryptographic hashing
    """
    
    def __init__(self, filename="database.csv", hash_file="file_hash.json"):
        self.filename = filename
        self.hash_file = hash_file
        self.baseline_hash = None
    
    def calculate_file_hash(self):
        """Calculate SHA-256 hash of the file"""
        sha256_hash = hashlib.sha256()
        
        try:
            with open(self.filename, "rb") as f:
                for byte_block in iter(lambda: f.read(4096), b""):
                    sha256_hash.update(byte_block)
            return sha256_hash.hexdigest()
        except FileNotFoundError:
            return None
    
    def create_sample_database(self):
        """Create a sample database.csv file"""
        data = {
            'ID': [1, 2, 3, 4, 5],
            'Name': ['John Doe', 'Jane Smith', 'Bob Johnson', 'Alice Brown', 'Charlie Wilson'],
            'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT'],
            'Salary': [50000, 60000, 75000, 55000, 65000],
            'Join_Date': ['2020-01-15', '2019-03-20', '2021-06-10', '2022-02-01', '2020-11-15']
        }
        
        df = pd.DataFrame(data)
        df.to_csv(self.filename, index=False)
        
        # Display styled output
        display(HTML("""
        <div style="background-color: #90EE90; padding: 10px; border-radius: 5px;">
            <h3 style="color: #006400;">✅ Sample Database Created Successfully!</h3>
        </div>
        """))
        
        print("\n📁 Initial database content:")
        display(df.style.set_properties(**{'background-color': '#f0f0f0', 
                                          'color': 'black', 
                                          'border': '1px solid black'}))
        return df
    
    def initialize_protection_profile(self):
        """Create and store baseline hash"""
        display(HTML("""
        <div style="background-color: #ADD8E6; padding: 10px; border-radius: 5px;">
            <h3>🔒 INITIALIZING PROTECTION PROFILE</h3>
        </div>
        """))
        
        current_hash = self.calculate_file_hash()
        
        if current_hash:
            self.baseline_hash = current_hash
            
            profile = {
                "filename": self.filename,
                "baseline_hash": current_hash,
                "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "algorithm": "SHA-256"
            }
            
            with open(self.hash_file, 'w') as f:
                json.dump(profile, f, indent=4)
            
            print(f"\n📁 File: {self.filename}")
            print(f"🔑 Baseline Hash: {current_hash}")
            print(f"⏰ Timestamp: {profile['timestamp']}")
            print(f"💾 Hash saved to: {self.hash_file}")
            
            return True
        return False
    
    def verify_integrity(self):
        """Verify file integrity against baseline"""
        display(HTML("""
        <div style="background-color: #FFF3CD; padding: 10px; border-radius: 5px;">
            <h3>🔍 VERIFYING FILE INTEGRITY</h3>
        </div>
        """))
        
        # Load baseline if not already loaded
        if not self.baseline_hash:
            try:
                with open(self.hash_file, 'r') as f:
                    profile = json.load(f)
                    self.baseline_hash = profile["baseline_hash"]
                    print(f"\n📋 Loaded baseline from: {profile['timestamp']}")
            except FileNotFoundError:
                display(HTML("""
                <div style="background-color: #FFB6C1; padding: 10px; border-radius: 5px;">
                    <h3 style="color: #8B0000;">❌ No protection profile found!</h3>
                    <p>Please initialize first (Run Cell 4).</p>
                </div>
                """))
                return
        
        current_hash = self.calculate_file_hash()
        
        if current_hash:
            print(f"\n📊 Hash Comparison:")
            print(f"   Baseline: {self.baseline_hash}")
            print(f"   Current:  {current_hash}")
            
            if current_hash == self.baseline_hash:
                display(HTML("""
                <div style="background-color: #90EE90; padding: 15px; border-radius: 5px;">
                    <h3 style="color: #006400;">✅✅✅ INTEGRITY CHECK PASSED ✅✅✅</h3>
                    <p style="font-size: 14px;">✓ File is in its original state</p>
                    <p style="font-size: 14px;">✓ No unauthorized modifications detected</p>
                </div>
                """))
                return True
            else:
                display(HTML("""
                <div style="background-color: #FFB6C1; padding: 15px; border-radius: 5px;">
                    <h3 style="color: #8B0000;">❌❌❌ INTEGRITY CHECK FAILED ❌❌❌</h3>
                    <p style="color: #8B0000; font-weight: bold;">⚠️ UNAUTHORIZED MODIFICATION DETECTED!</p>
                    <p>The file has been altered since baseline was created.</p>
                </div>
                """))
                
                # Show current file content
                print("\n📁 Current file content (after modification):")
                df = pd.read_csv(self.filename)
                display(df.style.set_properties(**{'background-color': '#FFE4E1', 
                                                  'color': 'black', 
                                                  'border': '1px solid red'}))
                return False
    
    def simulate_threat(self):
        """Simulate a threat by modifying the file"""
        display(HTML("""
        <div style="background-color: #FFE4E1; padding: 10px; border-radius: 5px;">
            <h3 style="color: #8B0000;">⚠️ SIMULATING THREAT - MODIFYING FILE</h3>
        </div>
        """))
        
        try:
            # Read the CSV file
            df = pd.read_csv(self.filename)
            
            print("\n📊 Before modification:")
            display(df.head(3).style.set_properties(**{'background-color': '#f0f0f0'}))
            
            # Modify a value (change salary)
            old_value = df.loc[0, 'Salary']
            employee_name = df.loc[0, 'Name']
            df.loc[0, 'Salary'] = 100000  # Changing from 50000 to 100000
            
            # Save the modified file
            df.to_csv(self.filename, index=False)
            
            print(f"\n✏️ Modified: Changed {employee_name}'s salary from ${old_value:,} to $100,000")
            
            print("\n📊 After modification:")
            display(df.head(3).style.set_properties(**{'background-color': '#FFE4E1'}))
            
            display(HTML("""
            <div style="background-color: #FFA07A; padding: 10px; border-radius: 5px;">
                <p style="font-weight: bold;">⚠️ THREAT SIMULATED: File has been modified!</p>
            </div>
            """))
            
        except FileNotFoundError:
            print("❌ Database file not found. Please create it first (Run Cell 3).")

# Create guardian instance
guardian = IntegrityGuardian()
print("✅ Integrity Guardian initialized!")
```
